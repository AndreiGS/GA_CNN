```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf
import pathlib


#from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
def get_data():
    dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
    ldata_dir = tf.keras.utils.get_file('flower_photos', origin=dataset_url, untar=True)
    ldata_dir = pathlib.Path(ldata_dir)

    return ldata_dir
```

```python
def save_model(model):
    model_json = model.to_json()
    with open("model.json", "w") as json_file:
        json_file.write(model_json)

    model.save_weights("model.h5")
    print("Saved model to disk")


def retrieve_model():
    print('retrieving model')

    json_file = open('model.json', 'r')
    loaded_model_json = json_file.read()
    json_file.close()
    loaded_model = tf.keras.models.model_from_json(loaded_model_json)
    # load weights into new model
    loaded_model.load_weights("model.h5")
    print("Loaded model from disk")

    return loaded_model
```

```python
data_dir = get_data()
image_count = len(list(data_dir.glob('*/*.jpg')))
print(image_count)
```

```python
img_height = 180
img_width = 180
batch_size = 32
train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size)
val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size)
class_names = train_ds.class_names
```

```python
data_augmentation = tf.keras.Sequential(
    [
        layers.RandomFlip("horizontal",
                          input_shape=(img_height,
                                       img_width,
                                       3)),
        layers.RandomRotation(0.1),
        layers.RandomZoom(0.1),
    ]
)

'''for better performance'''
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
num_classes = len(class_names)
from tensorflow.python.keras.callbacks import TensorBoard 
from datetime import date
```

```python
def generate_model(gene):
    model = Sequential()
    model.add(data_augmentation)
    model.add(layers.Rescaling(1./255, input_shape=(180,180,3)))
    for i in range (gene[0]):
        print('da')
        model.add(layers.Conv2D(32, 3, padding='same', activation='relu'))
        if gene[i+1] == 1:
            model.add(layers.MaxPooling2D())
    model.add(layers.Flatten())
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(num_classes))
    
    model.compile(optimizer='adam',
        loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
        metrics=['accuracy'])
    # model.summary()
    return model
```

```python
''' train '''
def train(model, epochs, nmb_of_generation, nmb_of_indiv):
    # log_dir = "logs/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
    today = date.today()

    # dd/mm/YY
    d1 = today.strftime("%d-%m-%Y")
    # tensorboard = TensorBoard(log_dir=)
    log = "logs/" + str(d1) + "/"+ str(nmb_of_generation) + "/" + str(nmb_of_indiv)
    tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log, histogram_freq=1)
    # loss_history = LossHistory()
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=epochs,
        callbacks=[tensorboard_callback]
    )
    return history.history['val_accuracy']
```

```python
def generate_initial_generation(nmb_indiv):
    arr = []
    for i in range(nmb_indiv):
        while True:
            nmb_layers = np.array([np.random.randint(2,6)])
            gene = np.random.randint(2,size=nmb_layers)
            poz_for_maxpooling = np.random.randint(0,nmb_layers)
            gene[poz_for_maxpooling] = 1
            gene = np.concatenate((nmb_layers, gene), axis=0)
            
            ok = True
            for el in arr:
                if np.array_equal(el, gene):
                    ok = False
            if ok:
                break
            
        arr.append(gene)
    arr = np.asarray(arr)
    print('Individuals:')
    print(arr)
    return arr
```

```python
import random
def update_generation(population, fitness, nmb_ind):
    # code cleanup, mutatia, prima antrenare  
    new_population = []
    idx = 0
    pop=[]
    for el in population:
        pop.append([el, fitness[idx]])
        idx=idx+1
        
    for i in range(int(nmb_ind/2)):
        parents = random.choices(pop, k=5) # ! vezi ca ii poate lua la fel
        parents = sorted(parents, key=lambda agent: agent[1], reverse=True)
        crossover_point = random.randint(0, min(parents[0][0][0], parents[1][0][0]))
        
        par1 = parents[0][0]
        par2 = parents[1][0]
        
        offspring1 = []
        offspring1.append(par1[0])
        offspring1 = offspring1 + par2[1:(crossover_point+2)] + par1[(crossover_point+2):]
        offspring2 = []
        offspring2.append(par2[0])
        offspring2 = offspring2 + par1[1:(crossover_point+2)] + par2[(crossover_point+2):]
        new_population.append(offspring1)
        new_population.append(offspring2)
        
    new_population = np.asarray(new_population)
    return new_population
```

```python
pop = np.array([[2, 0, 0],
               [3, 0, 1, 1],
               [4, 0, 1, 0, 1],
               [5, 0, 1, 1, 0, 1],
               [2, 1, 1],
               [4, 1, 0, 0, 1]])

fit = np.array([2, 6, 5, 2, 3, 7])
print(update_generation(pop, fit, 6))
```

```python
def train_by_GA(nmb_of_generations, nmb_ind_per_gen, nmb_of_epochs):
    ind = np.array(generate_initial_generation(nmb_ind_per_gen))
    print(ind)
    fitness = np.zeros(nmb_ind_per_gen)
    for gen in range(nmb_of_generations):
        idx_ind = 1
        for gene in ind:
            model = generate_model(gene)
            hist = train(model, nmb_of_epochs, gen+1, idx_ind)
            fitness[idx_ind-1] = hist[nmb_of_epochs-1]
            idx_ind = idx_ind+1
        print(fitness[idx_ind])
        # functia de update_generation care are nevoie de functia de selectie, crossing, mutation
```
